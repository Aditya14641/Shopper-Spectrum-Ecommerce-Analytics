# **Project Summary -**

The world of e-commerce is unfolding at an unimaginable scale, generating truckloads of transactional data every single day. Lying beneath this massive digital footprint, however, is a wealth of hidden information about customer buying patterns, evolving market demands, and complicated product interactions. In today’s hyper-competitive digital space, intuition isn’t enough – analysis of these historical transaction patterns is paramount for segmentation of customers into logically coherent groups and proactive recommendations for relevant products. Beyond enhancing user experience through personalization, this process actively boosts revenue by driving cross-selling and customer retention initiatives. The aim of this capstone project is to analyze the intricate transaction logs of an online retail business to reveal hidden purchasing patterns, categorize customers into logical segments, and facilitate seamless cross-selling of products to maximize their Customer Lifetime Value (CLV).

The project is structured systematically into three main stages: Exploratory Data Analysis (EDA), Machine Learning Modeling and Application Deployment. The first stage, EDA, involves meticulously cleansing the raw, unstructured transaction data to ensure data integrity. This thorough pre-processing includes removal of records with missing customer IDs, as transactions can't contribute to behavior profiling if they can't be linked to specific individuals. Additionally, transactions representing canceled orders (negative quantity) and erroneous zero price records were eliminated to prevent skewed financial metrics. A new and vital ‘Total Amount’ metric is engineered by taking product quantity multiplied by unit price to offer a direct insight into total generated revenue. Once data is clean, comprehensive visuals utilizing Python’s data visualization libraries are constructed to illustrate macro-level business trends. This includes charting best-selling products by quantity, top-grossing countries for international revenue, and distributions in customer buying habits, such as busiest purchasing times and seasonal sales peaks.

Moving into the Machine Learning phase, the project leverages two separate, unsupervised learning methods to uncover actionable insights. Firstly, a Recency, Frequency and Monetary (RFM) method is mathematically constructed for each unique customer. This process involves quantifying: how recently a customer made a purchase (Recency), how frequently they buy from us (Frequency), and the total monetary value they have spent over their customer lifetime (Monetary). Given the drastically different scales of these values, the resulting RFM matrix is standardized using a data scaler. This normalized data is subsequently used in conjunction with a K-Means Clustering algorithm to automatically partition and group the customer base into distinct, coherent segments (e.g., High-value VIPs, Frequent loyalists, One-time purchasers, lapsed or At-risk customers, and new customers). The number of resulting customer segments is systematically evaluated using quantitative metrics, such as Silhouette Score and the Elbow Method, to validate mathematical integrity of the resulting groups.

Secondly, the project address customer product discovery by implementing an Item-Based Collaborative Filtering recommendation engine. This model avoids generic popularity based recommendations by digging into deeper, co-purchase historical data. Data is first transformed into a Customer-Item pivot matrix, where a sparsity exists representing purchasing relationships between users and products. Based on the relative frequency that two items are included within the same shopping cart, a Cosine Similarity score can be computed for all items within the product catalog. This allows the model to take any given product in the store and instantly recommend the top 5 most similar items, which is perfect for the “customers also bought” and targeted cross-selling scenarios.

The third and final stage involves bridging the gap between pure data science and valuable business intelligence through application deployment. The developed K-Means model, along with the standard data scaler and the Cosine Similarity matrix, are saved as serialized objects and incorporated into an interactive Streamlit Web Application. This intuitive UI enables non-technical stakeholders such as marketing and sales teams to seamlessly integrate the machine learning outputs into their workflows. They can use the application to input live customer RFM scores and receive their predicted customer segment instantaneously for highly targeted marketing efforts, and they can select products from a dropdown menu for on-the-fly product recommendation lists. Ultimately, this comprehensive project successfully demonstrates the transformation of raw e-commerce data into an automated and scalable system that personalizes the customer experience and maximizes operational ROI.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Drop rows where CustomerID is missing, as we cannot track behavior without it
df.dropna(subset=['CustomerID'], inplace=True)

# Remove canceled orders (Quantity <= 0) and free items/errors (UnitPrice <= 0)
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Convert InvoiceDate to a proper datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create a new feature for the total transaction amount
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']

# Extract useful time-based features
df['Month'] = df['InvoiceDate'].dt.to_period('M')
df['Hour'] = df['InvoiceDate'].dt.hour
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()

# Display the clean dataset shape
print(f"Cleaned Dataset Shape: {df.shape}")
````

````python
# --- PREPARING RFM FOR EDA ---
import datetime as dt
latest_date = df['InvoiceDate'].max() + dt.timedelta(days=1)
rfm_eda = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalAmount': 'sum'
}).rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'TotalAmount': 'Monetary'})
rfm_eda = rfm_eda[rfm_eda['Monetary'] > 0]
````
